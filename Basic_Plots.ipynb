```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Line Chart

```python
temp = [24,34,28,45,34,22,38]
days = [1,2,3,4,5,6,7]
plt.plot(days,temp, marker = 'o', ms = 10, mfc = 'k', mec = 'black', color = 'blue', lw = 3, ls = '-.')#ms = marker size, mfc = marker fill color, mec = marker edge color
plt.xlabel("Days", size = 15, color = 'g')#g or green
plt.ylabel("Tempterature", size = 15, color = 'g')#b or blue
plt.title("My First Plot", size = 15, color = 'darkblue')
plt.show()
```

```python
max_temp = [45,47,38,34,25,41,48]
avg_temp = [30,32,29,25,15,33,32]
min_temp = [25,29,20,21,10,26,29]
plt.plot(days, max_temp, label = 'MAX')
plt.plot(days, avg_temp, label = 'AVG', ls = '--')
plt.plot(days, min_temp, label = 'MIN')
plt.legend(loc = 'best')
plt.show()
```

```python
year = [2015,2016,2017,2018,2019,2020]
income = [45,67,36,90,98,50]
profit = [40,50,28,80,70,40]
plt.bar(year, income, color = 'blue', label = 'Income')
plt.bar(year, profit, color = 'orange', label = 'Profit')
plt.xlabel("Year", size = 15, color = 'g')#g or green
plt.ylabel("Income", size = 15, color = 'g')#b or blue
plt.legend()
plt.show()
```
