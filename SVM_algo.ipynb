```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.datasets import make_circles
```

```python
data = pd.read_csv('Datasets/SVM.csv')
```

```python
data.head()
```

```python
data.tail()
```

```python
x = data[['Flour', 'Sugar']]
y = data['Type']
```

```python
sns.lmplot('Flour', 'Sugar', hue = 'Type',data = data, fit_reg = False)
plt.show()
```

```python
le = len(data)
```

```python
ind = np.random.permutation(le)
```

```python
x = np.array(x)
y = np.array(y)
```

```python
x = x[ind]
y = y[ind]
```

```python
model = SVC(kernel = 'linear')
mt = model.fit(x,y)
```

```python
model.predict([[42,20]])
```

```python
ypred = model.predict(x)
```

```python
accuracy_score(y, ypred) * 100
```

```python
sv = model.support_vectors_
```

```python
plt.scatter(x[:,0],x[:,1])
plt.scatter(sv[:,0],sv[:,1])
plt.show
```
