## Filtering Based Method

```python
import pandas as pd
import warnings
```

```python
warnings.filterwarnings("ignore", category=RuntimeWarning) 
```

```python
Movies = pd.read_csv('Datasets/movies.csv')
Ratings = pd.read_csv('Datasets/ratings.csv')
```

```python
Movies.head()
```

```python
Ratings.head()
```

```python
df = pd.merge(Movies,Ratings, on = 'movieId')
```

```python
vv = df['title'].value_counts
```

```python
df.head()
```

```python
def Lower_case_converter(title):
    return title.lower()
```

```python
df['title'] = df['title'].apply(Lower_case_converter)
```

```python
df.head()
```

```python
df['title'] = df['title'].str.replace(r"\s*\(.*\)","") #removing year from the front of the movie name
```

```python
group = df.groupby('title')['rating'].mean().sort_values(ascending = False)
```

```python
df.groupby('title')['rating'].count().sort_values(ascending = False)
```

```python
Newdf = pd.DataFrame(df.groupby('title')['rating'].mean())
```

```python
Newdf['NumOfRating'] = df.groupby('title')['rating'].count()
```

```python
Newdf
```

```python
Movie_Title = df.pivot_table(index = 'userId', columns = 'title', values = 'rating')
```

```python
Movie_Title
```

```python
Movie_Title['toy story']
```

```python
def Rec():
    MovieName = input("Enter Movie Name: ").lower()
    try:
        user_ratings = Movie_Title[MovieName]
        similar_movies = Movie_Title.corrwith(user_ratings)
        corr_movie = pd.DataFrame(similar_movies, columns = ['Correlation'])
        corr_movie.dropna(inplace = True)
        corr_movie = corr_movie.join(Newdf['NumOfRating'])
        sugg = corr_movie[corr_movie['NumOfRating']>100].sort_values(['Correlation'], ascending = False)
        top_5 = sugg[:10].index
        print("Similar Movies:")
        for i in top_5:
            print(i)
    except:
        print('No suggestions')
```

```python
Rec()
```

 ## Content Based Method

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
Moviedata = pd.read_csv('Datasets/data.csv')
```

```python
Moviedata.shape
```

```python
Moviedata.head()
```

```python
Moviedata.keys()
```

```python
Moviedata.info()
```

```python
features = ['genres', 'keywords', 'cast', 'director']
```

```python
for i in features:
    Moviedata[i] = Moviedata[i].fillna("")
```

```python
def CreateContent(row):
    content = ""
    for i in features:
        content += row[i]
        content += " "
    return content    

def LowerCase(title):
    return title.lower()
    
```

```python
Moviedata['original_title'] = Moviedata['original_title'].apply(LowerCase)
```

```python
Moviedata['movie_content'] = Moviedata.apply(CreateContent, axis = 1)
```

```python
Moviedata['movie_content']
```

```python
Moviedata['original_title']
```

```python
cv = CountVectorizer()
count = cv.fit_transform(Moviedata['movie_content'])
```

```python
count.toarray().shape
```

```python
cv.get_feature_names()
```

```python
row_similarity = cosine_similarity(count)
```

```python
row_similarity
```

```python
def ContentBasedRecomm():
    movie_name = input("Enter the movie name: ").lower()
    try:
        m_index = Moviedata[Moviedata['original_title'] == movie_name]['index'].values[0]
        similar_word_value = row_similarity[m_index]
        pairing_value = list(enumerate(similar_word_value))
        similar_movie = sorted(pairing_value, key = lambda x:x[1], reverse = True)
        num = 5
        print("Top 5 Similar Movies..")
        for i in range(num):
            index = similar_movie[i][0]
            movie = Moviedata[index:index+1]['original_title'].values[0]
            print(i+1, movie)    
    except IndexError:
        print("No suggestions..")
```

```python
ContentBasedRecomm()
```
