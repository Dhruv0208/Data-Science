```python
import pandas as  pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler, Binarizer
from word2number import w2n
```

```python
df = pd.read_csv('Datasets/income.csv')
```

```python
df
```

## implementing  MinMaxScaler using formula

```python
MIN = df['Age'].min()
MAX = df['Age'].max()
```

```python
diffe = MAX - MIN
```

```python
df['NewAge'] = (df['Age'] - MIN)/diffe
```

```python
df
```

## MinMaxSclaer using prebuilt

```python
a = MinMaxScaler()
```

```python
ages = a.fit_transform(df[['Age']])
```

```python
ages
```

## Standard scaler using formula

```python
mm = df['Income($)'].mean()
st = df['Income($)'].std()
```

```python
df['NewIncome'] = (df['Income($)'] - mm)/st
```

```python
df
```

## standardScaler using prebuilt

```python
std_scaler = StandardScaler()
Income = std_scaler.fit_transform(df[['Income($)']])
```
